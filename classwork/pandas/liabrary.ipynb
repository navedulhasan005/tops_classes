```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('diamonds.csv')
df
```

```python
df1 = df.groupby('cut')['price'].sum().reset_index()
df1
```

```python
plt.figure(figsize=(10,5))
plt.plot(df1['cut'],df1['price'],marker = "o",color = "blue",linestyle = "-")
plt.title("Cut Wise Price")
plt.xlabel("Cut of diamond")
plt.ylabel("Total Price")
plt.grid(True)
```

```python
sns.lineplot(data=df1,x="cut",y="price",marker="o",color="red")
plt.title("Cut Wise Price")
plt.grid(True)
```

```python
df2 = df.groupby('carat')['price'].mean().reset_index()
df2
```

```python
plt.figure(figsize=(10,5))
plt.scatter(df2['carat'],df2['price'],marker = "o",color = "blue",alpha=0.4)
plt.title("carat Wise Avrage Price")
plt.xlabel("carat of diamond")
plt.ylabel("Total Avrage Price")
plt.grid(True)
```

```python
sns.scatterplot(data=df2,x="carat",y="price",alpha=0.4,color="blue")
```

```python
plt.figure(figsize=(10,5))
plt.bar(df1['cut'],df1['price'],color = "blue",linestyle = "-")
plt.title("Cut Wise Price")
plt.xlabel("Cut of diamond")
plt.ylabel("Total Price")
```

```python
sns.barplot(data=df1,x="cut",y="price",color="red")
plt.title("Cut Wise Price")
plt.grid(False)
```

```python
df3 = df.groupby(['clarity','cut'])['price'].sum().reset_index()
df3
```

```python
sns.barplot(data=df3,x='cut',y="price",hue="clarity",palette="viridis")
plt.title("Clarity and Cut wise price")
```
