### merge and join

```python
import pandas as pd

df = pd.read_csv('diamonds.csv')

df
```

```python
df1 = pd.DataFrame({'ID':[1,2,3],
                   'Name':['Alice','John','Bob']})
df1
```

```python
df2 = pd.DataFrame({'ID':[2,3,4],
                   'Score':[50,40,70]})
df2
```

```python
merge_data = pd.merge(df1,df2, on='ID',how = 'inner')
merge_data
```
