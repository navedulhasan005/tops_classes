### merge and join

```python
df1 = pd.DataFrame({'ID':[1,2,3],
                   'Name':['Alice','John','Bob']})
df1
```

```python
df2 = pd.DataFrame({'ID':[2,3,4],
                   'Score':[50,40,70]})
df2
```

```python
#merge merge table using column
merge_data = pd.merge(df1,df2, on='ID',how = 'left')
# merge_data = pd.merge(df1,df2, on='ID',how = 'outer')
# merge_data = pd.merge(df1,df2, on='ID',how = 'left')
merge_data
```

```python
#join join table using index
data1 = pd.DataFrame({'Name':['Alice','John','Bob']},index =[1,2,3])
data2 = pd.DataFrame({'Score':[50,40,70]},index = [2,3,4])
join_data = data1.join(data2, how='outer')
# join_data = data1.join(data2, how='left')
# join_data = data1.join(data2, how='inner')

join_data
```

### Where Function

```python
import pandas as pd
import numpy as np

df = pd.read_csv('diamonds.csv')

df
```

```python
# df["New"] = np.where(df['cut']=="Good", "Yes","No")
# df["New"] = np.where((df['cut']=="Good")|(df['cut']=="Ideal"), "Yes","No")
df["New"] = np.where(df['color'].isin(['D','E']), "Yes","No")

df.head(10)
```

```python
# df["New_price"] = np.where(df['color'].isin(['I','J']),df['price']*2,df['price'])
# df["New_price"] = np.where((df['color'] == 'D') & (df['cut'] == 'Premium') ,df['price']*2,df['price'])
df["New_price"] = np.where((df['color'] == 'D') & (df['cut'] == 'Premium') ,df['price']*2,df['price'])
df
```

```python
df['Depth_Between'] = np.where(df['depth']<60,"Less then 60",
                      np.where(df['depth']<65,"60-65",
                      np.where(df['depth']<70,"65-70","More then 70")))
df.loc[100:500]
```
